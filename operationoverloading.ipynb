```python
class Student:
    def __init__(self,marks):
        self.marks = marks

    def __add__(self,others):
        return self.marks+others.marks
s1 = Student(80)
s2 = Student(75)
print(s1+s2)
```

```python
class Student:
    def __init__(self,marks):
        self.marks = marks

    def __sub__(self,others):
        return self.marks-others.marks
s1 = Student(80)
s2 = Student(75)
print(s1-s2)
```
